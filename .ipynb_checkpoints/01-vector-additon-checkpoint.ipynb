```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the results from jsonl file
data = []
with open('01-vector-additon-perf.jsonl', 'r') as f:
    for line in f:
        if line.strip():
            data.append(json.loads(line))

# Filter out errors and extract data
results = []
for entry in data:
    if 'error' not in entry and 'gpu_perf' in entry and 'cpu_perf' in entry:
        experiment = entry['config'].get('experiment', {})
        experiment_name = experiment.get('name', 'unknown') if experiment else 'unknown'
        
        results.append({
            'vector_size': entry['config']['vector_size'],
            'threads_per_block': entry['config']['threads_per_block'],
            'gpu_time_us': entry['gpu_perf']['time_us'],
            'cpu_time_us': entry['cpu_perf']['time_us'],
            'gpu_throughput': entry['gpu_perf']['throughput_gib_s'],
            'cpu_throughput': entry['cpu_perf']['throughput_gib_s'],
            'experiment': experiment_name
        })

df = pd.DataFrame(results)
print(f"Loaded {len(df)} successful benchmark results")
print(f"Experiments found: {df['experiment'].unique()}")
df.head()
```

```python
# Filter data for vector_size_sweep experiment
df_main = df[df['experiment'] == 'vector_size_sweep'].copy()

if len(df_main) == 0:
    # Fallback for old data without experiment field
    df_main = df[df['threads_per_block'] == 256].copy()
    max_size_idx = df_main['vector_size'].idxmax()
    df_main = df_main.loc[:max_size_idx]

# Plot CPU vs GPU execution time
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df_main['vector_size'], df_main['gpu_time_us'], 'o-', label='GPU', linewidth=2, markersize=6)
ax.plot(df_main['vector_size'], df_main['cpu_time_us'], 's-', label='CPU', linewidth=2, markersize=6)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Vector Size (log scale)', fontsize=12)
ax.set_ylabel('Execution Time (s, log scale)', fontsize=12)
ax.set_title('CPU vs GPU Execution Time (Vector Size Sweep)', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()

# Print speedup statistics
df_main['speedup'] = df_main['cpu_time_us'] / df_main['gpu_time_us']
print(f"\nSpeedup Statistics:")
print(f"  Min speedup: {df_main['speedup'].min():.2f}x")
print(f"  Max speedup: {df_main['speedup'].max():.2f}x")
print(f"  Mean speedup: {df_main['speedup'].mean():.2f}x")
print(f"\nSpeedup at largest size ({df_main['vector_size'].max()} elements): {df_main['speedup'].iloc[-1]:.2f}x")
```

```python
# Plot throughput comparison (vector size sweep only)
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df_main['vector_size'], df_main['gpu_throughput'], 'o-', label='GPU', linewidth=2, markersize=6, color='green')
ax.plot(df_main['vector_size'], df_main['cpu_throughput'], 's-', label='CPU', linewidth=2, markersize=6, color='orange')

ax.set_xscale('log')
ax.set_xlabel('Vector Size (log scale)', fontsize=12)
ax.set_ylabel('Throughput (GiB/s)', fontsize=12)
ax.set_title('CPU vs GPU Throughput (Vector Size Sweep)', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Plot effect of threads_per_block on GPU performance
df_threads = df[df['experiment'] == 'threads_sweep'].copy()

if len(df_threads) == 0:
    # Fallback for old data
    df_threads = df[df['vector_size'] == 1000000].copy()

if len(df_threads) > 0:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    # Execution time vs threads
    ax1.plot(df_threads['threads_per_block'], df_threads['gpu_time_us'], 'o-', linewidth=2, markersize=8)
    ax1.set_xlabel('Threads per Block', fontsize=12)
    ax1.set_ylabel('GPU Execution Time (s)', fontsize=12)
    ax1.set_title('GPU Time vs Threads per Block', fontsize=13)
    ax1.grid(True, alpha=0.3)

    # Throughput vs threads
    ax2.plot(df_threads['threads_per_block'], df_threads['gpu_throughput'], 'o-', linewidth=2, markersize=8, color='green')
    ax2.set_xlabel('Threads per Block', fontsize=12)
    ax2.set_ylabel('GPU Throughput (GiB/s)', fontsize=12)
    ax2.set_title('GPU Throughput vs Threads per Block', fontsize=13)
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("\nThreads per Block Impact:")
    print(df_threads[['threads_per_block', 'gpu_time_us', 'gpu_throughput']].to_string(index=False))
else:
    print("No threads_sweep experiment data found")
```

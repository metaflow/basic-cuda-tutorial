```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the results from jsonl file
data = []
with open('01-vector-additon-perf.jsonl', 'r') as f:
    for line in f:
        if line.strip():
            data.append(json.loads(line))

# Filter out errors and extract data
results = []
for entry in data:
    if 'error' not in entry and 'gpu_perf' in entry:
        # Handle new hierarchical config format
        runtime_config = entry['config'].get('runtime', entry['config'])
        experiment = runtime_config.get('experiment', {})
        experiment_name = experiment.get('name', 'unknown') if experiment else 'unknown'
        
        results.append({
            'vector_size': runtime_config['vector_size'],
            'threads_per_block': runtime_config['threads_per_block'],
            'gpu_time_us': entry['gpu_perf']['time_us'],
            'cpu_time_us': entry['cpu_perf']['time_us'] if 'cpu_perf' in entry else np.nan,
            'gpu_throughput': entry['gpu_perf']['throughput_gib_s'],
            'cpu_throughput': entry['cpu_perf']['throughput_gib_s'] if 'cpu_perf' in entry else np.nan,
            'experiment': experiment_name,
            'full_config': entry['config']  # Store the complete config
        })

df = pd.DataFrame(results)
print(f"Loaded {len(df)} successful benchmark results")
print(f"Experiments found: {df['experiment'].unique()}")
df.head()
```

```python
# Filter data for vector_size_sweep experiment
df_main = df[df['experiment'] == 'vector_size_sweep'].copy()
# Sort by vector size for proper line plotting
df_main = df_main.sort_values('vector_size')

# Plot CPU vs GPU execution time
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df_main['vector_size'], df_main['gpu_time_us'], 'o-', label='GPU', linewidth=2, markersize=6)
ax.plot(df_main['vector_size'], df_main['cpu_time_us'], 's-', label='CPU', linewidth=2, markersize=6)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Vector Size (log scale)', fontsize=12)
ax.set_ylabel('Execution Time (s, log scale)', fontsize=12)
ax.set_title('CPU vs GPU Execution Time (Vector Size Sweep)', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()

# Print speedup statistics
df_main['speedup'] = df_main['cpu_time_us'] / df_main['gpu_time_us']
print(f"\nSpeedup Statistics:")
print(f"  Min speedup: {df_main['speedup'].min():.2f}x")
print(f"  Max speedup: {df_main['speedup'].max():.2f}x")
print(f"  Mean speedup: {df_main['speedup'].mean():.2f}x")
print(f"\nSpeedup at largest size ({df_main['vector_size'].max()} elements): {df_main['speedup'].iloc[-1]:.2f}x")
```

```python
import numpy as np
import copy


def generate_midpoint_config(config1, config2, x_col, log_x=True):
    """
    Generate a single midpoint config between two configs.

    Args:
        config1: First config (dict with 'full_config' field and x_col value)
        config2: Second config (dict with 'full_config' field and x_col value)
        x_col: Column name for the parameter to interpolate
        log_x: Whether to use log spacing (True) or linear spacing (False)

    Returns:
        A new config dict with the interpolated x_col value
    """
    x1 = config1[x_col]
    x2 = config2[x_col]

    # Calculate midpoint based on spacing mode
    if log_x:
        midpoint_x = int(np.exp((np.log(x1) + np.log(x2)) / 2))
    else:
        midpoint_x = int((x1 + x2) / 2)

    # Use first config as template
    new_config = copy.deepcopy(config1['full_config'])
    new_config['runtime'][x_col] = midpoint_x

    return new_config


def find_slope_changes(df_exp, x_col, y_col, log_x=True):
    """
    Find points with biggest changes in slope (second derivative).
    Weights slope changes by the interval length to prioritize changes over larger ranges.
    Returns points with interpolation configs already generated.

    Args:
        df_exp: DataFrame with experiment data (must be sorted by x_col)
        x_col: Column name for x-axis
        y_col: Column name for y-axis
        log_x: Whether to use log scale for x-axis

    Returns:
        List of dicts with slope change info and generated interpolation configs
    """
    if len(df_exp) < 3:
        return []

    # Get x and y values (original scale for interval calculation)
    x_orig = df_exp[x_col].values
    y = df_exp[y_col].values

    # Apply log to x for slope calculation if requested
    x = np.log(x_orig) if log_x else x_orig

    # Calculate first derivative (slope between consecutive points)
    slopes = np.diff(y) / np.diff(x)

    # Calculate second derivative (change in slope)
    slope_changes = np.abs(np.diff(slopes))

    # Calculate interval lengths (in the appropriate scale)
    # For each middle point, the interval is from x[i-1] to x[i+1]
    if log_x:
        # In log scale: log(x_right) - log(x_left) = log(x_right / x_left)
        interval_lengths = np.log(x_orig[2:] / x_orig[:-2])
    else:
        # In linear scale: x_right - x_left
        interval_lengths = x_orig[2:] - x_orig[:-2]

    # Weight slope changes by interval length
    weighted_slope_changes = slope_changes * interval_lengths

    # Find indices sorted by weighted slope change magnitude
    # slope_changes[i] corresponds to point i+1 (middle point of three consecutive points)
    sorted_indices = np.argsort(weighted_slope_changes)[::-1]

    results = []
    for idx in sorted_indices:
        middle_idx = idx + 1  # Index in original dataframe

        # Get the three points involved
        config_before = df_exp.iloc[middle_idx - 1].to_dict()
        config_middle = df_exp.iloc[middle_idx].to_dict()
        config_after = df_exp.iloc[middle_idx + 1].to_dict()

        # Generate interpolation configs (two midpoints)
        interp_configs = [
            generate_midpoint_config(config_before, config_middle, x_col, log_x),
            generate_midpoint_config(config_middle, config_after, x_col, log_x)
        ]

        results.append({
            'middle_index': middle_idx,
            'middle_value': {x_col: config_middle[x_col], y_col: config_middle[y_col]},
            'slope_change': slope_changes[idx],
            'interval_length': interval_lengths[idx],
            'weighted_slope_change': weighted_slope_changes[idx],
            'config_before': config_before,
            'config_middle': config_middle,
            'config_after': config_after,
            'interpolation_configs': interp_configs
        })

    return results


def find_biggest_step(df_exp, x_col, y_col, log_x=True):
    """
    Find the pair of consecutive points with the biggest absolute step in y value.
    Returns the pair with a single interpolation config between them.

    Args:
        df_exp: DataFrame with experiment data (must be sorted by x_col)
        x_col: Column name for x-axis
        y_col: Column name for y-axis
        log_x: Whether to use log spacing for interpolation

    Returns:
        List of dicts with step info and generated interpolation config (one midpoint)
    """
    if len(df_exp) < 2:
        return []

    y = df_exp[y_col].values

    # Calculate absolute step between consecutive points
    steps = np.abs(np.diff(y))

    # Find indices sorted by step magnitude
    sorted_indices = np.argsort(steps)[::-1]

    results = []
    for idx in sorted_indices:
        # idx is the index of the first point in the pair
        config1 = df_exp.iloc[idx].to_dict()
        config2 = df_exp.iloc[idx + 1].to_dict()

        # Generate single interpolation config (midpoint between the two)
        interp_config = generate_midpoint_config(config1, config2, x_col, log_x)

        results.append({
            'pair_index': idx,
            'pair_values': {
                x_col: (config1[x_col], config2[x_col]),
                y_col: (config1[y_col], config2[y_col])
            },
            'step_magnitude': steps[idx],
            'config1': config1,
            'config2': config2,
            'interpolation_configs': [interp_config]
        })

    return results
```

```python
# Example usage: Find interesting points in vector_size_sweep experiment
experiment = 'threads_sweep'
x_value = 'threads_per_block'
use_log_x = False
y_value = 'gpu_throughput'

df_analysis = df[df['experiment'] == experiment].copy()
df_analysis = df_analysis.sort_values(x_value)

# Filter out NaN values in gpu_throughput
df_analysis = df_analysis[df_analysis[y_value].notna()]

print("=" * 80)
print(f"ANALYZING {experiment.upper()} FOR INTERESTING TRANSITIONS")
print("=" * 80)

if len(df_analysis) < 3:
    print("Not enough data points for analysis")
else:
    all_interpolation_configs = []
    
    # Method 1: Find slope changes (second derivative - acceleration/deceleration)
    # Now weighted by interval length
    slope_changes = find_slope_changes(df_analysis, x_value, y_value, log_x=use_log_x)
    print(f"\n1. SLOPE CHANGES (Second Derivative, weighted by interval)")
    print(f"   Found {len(slope_changes)} points with slope changes")
    print(f"   Using {'log' if use_log_x else 'linear'} spacing\n")
    
    for i, point in enumerate(slope_changes[:3]):
        print(f"   #{i+1}. Point at {x_value}={point['middle_value'][x_value]:,}")
        print(f"       {y_value}: {point['middle_value'][y_value]:.2f}")
        print(f"       Slope change: {point['slope_change']:.4f}")
        print(f"       Interval length: {point['interval_length']:.4f}")
        print(f"       Weighted score: {point['weighted_slope_change']:.4f}")
        print(f"       X: {point['config_before'][x_value]:,}  {point['config_middle'][x_value]:,}  {point['config_after'][x_value]:,}")
        print(f"       Y: {point['config_before'][y_value]:.2f}  {point['config_middle'][y_value]:.2f}  {point['config_after'][y_value]:.2f}")
        
        # Configs are already generated
        interp_configs = point['interpolation_configs']
        all_interpolation_configs.extend(interp_configs)
        
        print(f"        Interpolation points: {interp_configs[0]['runtime'][x_value]:,}, {interp_configs[1]['runtime'][x_value]:,}")
        print()

    # Method 2: Find biggest steps (first derivative - sudden jumps)
    biggest_steps = find_biggest_step(df_analysis, x_value, y_value, log_x=use_log_x)
    print(f"\n2. BIGGEST STEPS (Absolute Y-value jumps)")
    print(f"   Found {len(biggest_steps)} pairs sorted by step size")
    print(f"   Using {'log' if use_log_x else 'linear'} spacing\n")
    
    for i, point in enumerate(biggest_steps[:3]):
        x1, x2 = point['pair_values'][x_value]
        y1, y2 = point['pair_values'][y_value]
        print(f"   #{i+1}. Between {x_value} {x1:,} and {x2:,}")
        print(f"       {y_value}: {y1:.2f}  {y2:.2f}")
        print(f"       Step magnitude: {point['step_magnitude']:.4f}")
        
        # Config is already generated (single midpoint)
        interp_config = point['interpolation_configs'][0]
        all_interpolation_configs.append(interp_config)
        
        print(f"        Interpolation point: {interp_config['runtime'][x_value]:,}")
        print()
    
    # Deduplicate configs
    if all_interpolation_configs:
        # Get existing x values from the dataframe
        existing_x_values = set(df_analysis[x_value].values)
        
        unique_configs = []
        seen_configs = set()
        duplicates_with_existing = []
        duplicates_with_generated = []
        
        for config in all_interpolation_configs:
            x_val = config['runtime'][x_value]
            
            # Check if this x value already exists in the data
            if x_val in existing_x_values:
                duplicates_with_existing.append(x_val)
                continue
            
            # Check if we've already generated this config
            config_key = json.dumps(config, sort_keys=True)
            if config_key in seen_configs:
                duplicates_with_generated.append(x_val)
                continue
            
            seen_configs.add(config_key)
            unique_configs.append(config)
        
        print("=" * 80)
        print(f"GENERATED CONFIGS")
        print("=" * 80)
        print(f"Total generated: {len(all_interpolation_configs)}")
        print(f"Duplicates with existing data: {len(duplicates_with_existing)} {sorted(set(duplicates_with_existing))}")
        print(f"Duplicates within generated: {len(duplicates_with_generated)}")
        print(f"Unique new configs: {len(unique_configs)}\n")
        for config in unique_configs:
            print(json.dumps(config))
```
